# BuildersVault Social Services Hackathon, Track 2 Quickstart

**Food Security Delivery Operations**

This notebook matches the current Track 2 generator output: 500 clients, 10,000 delivery requests, 300 routes, about 3,500 route stops, and 150 inventory items.

```python
# Imports and config
import sys
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

pd.set_option('display.max_columns', 80)
pd.set_option('display.width', 180)
plt.rcParams['figure.dpi'] = 110

repo_root = Path.cwd()
candidate_dirs = [
    repo_root / 'tracks' / 'food-security-delivery' / 'data' / 'raw',
    repo_root.parent / 'data' / 'raw',
]
DATA_DIR = next((path for path in candidate_dirs if path.exists()), None)

if DATA_DIR is None:
    print('Data directory not found.')
    print('Run the generator from repo root first:')
    print('    python3 tracks/food-security-delivery/generator/generate.py')
    sys.exit(0)

print(f'Data directory: {DATA_DIR.resolve()}')
print(f'Files present: {sorted(p.name for p in DATA_DIR.glob("*.parquet"))}')
```

```python
# Load all 9 Parquet files
clients = pd.read_parquet(DATA_DIR / 'clients.parquet')
drivers = pd.read_parquet(DATA_DIR / 'drivers.parquet')
vehicles = pd.read_parquet(DATA_DIR / 'vehicles.parquet')
depots = pd.read_parquet(DATA_DIR / 'depots.parquet')
routes = pd.read_parquet(DATA_DIR / 'routes.parquet')
requests = pd.read_parquet(DATA_DIR / 'delivery_requests.parquet')
stops = pd.read_parquet(DATA_DIR / 'route_stops.parquet')
items = pd.read_parquet(DATA_DIR / 'inventory_items.parquet')
request_items = pd.read_parquet(DATA_DIR / 'delivery_request_items.parquet')

tables = {
    'clients': clients,
    'drivers': drivers,
    'vehicles': vehicles,
    'depots': depots,
    'routes': routes,
    'delivery_requests': requests,
    'route_stops': stops,
    'inventory_items': items,
    'delivery_request_items': request_items,
}

for name, df in tables.items():
    print(f'{name:28s} shape={df.shape}')
```

```python
# Null summary across all 9 tables
summary_rows = []
for name, df in tables.items():
    nulls = df.isna().sum()
    pct_nulls = (nulls / len(df) * 100).round(2)
    null_cols = pct_nulls[pct_nulls > 0]
    summary_rows.append({
        'table': name,
        'rows': len(df),
        'cols': df.shape[1],
        'columns_with_nulls': len(null_cols),
        'max_null_pct': float(null_cols.max()) if len(null_cols) else 0.0,
        'worst_null_column': null_cols.idxmax() if len(null_cols) else None,
    })

null_summary = pd.DataFrame(summary_rows)
null_summary
```

```python
# Schema validation against the current generator output
expected_schemas = {
    'clients': {
        'client_id': 'object', 'first_name': 'object', 'last_name': 'object',
        'lat': 'float64', 'lng': 'float64', 'home_depot_id': 'object',
        'enrolment_status': 'object', 'food_security_level': 'object',
        'language_primary': 'object', 'mobility_wheelchair': 'bool',
        'requires_two_person_team': 'bool',
    },
    'drivers': {
        'driver_id': 'object', 'first_name': 'object', 'last_name': 'object',
        'role_type': 'object', 'home_base_depot_id': 'object', 'vehicle_id': 'object',
        'can_handle_wheelchair': 'bool', 'pet_allergy_flag': 'bool',
        'language_skills': 'object', 'max_hours': 'int64',
    },
    'vehicles': {
        'vehicle_id': 'object', 'type': 'object', 'capacity_meals': 'int64',
        'capacity_weight_kg': 'float64', 'refrigerated': 'bool', 'wheelchair_lift': 'bool',
    },
    'depots': {
        'depot_id': 'object', 'name': 'object', 'lat': 'float64', 'lng': 'float64',
    },
    'routes': {
        'route_id': 'object', 'service_date': 'object', 'driver_id': 'object',
        'vehicle_id': 'object', 'planned_stops': 'int64', 'actual_stops': 'int64',
        'route_status': 'object', 'on_time_rate': 'float64',
    },
    'delivery_requests': {
        'request_id': 'object', 'client_id': 'object', 'scheduled_date': 'object',
        'time_window_start': 'object', 'time_window_end': 'object', 'program_type': 'object',
        'priority_level': 'object', 'cold_chain_required': 'bool',
    },
    'route_stops': {
        'route_stop_id': 'object', 'route_id': 'object', 'request_id': 'object',
        'client_id': 'object', 'sequence_index': 'int64', 'actual_arrival': 'datetime64',
        'status': 'object', 'failure_reason': 'object',
    },
    'inventory_items': {
        'item_id': 'object', 'name': 'object', 'allergen_flags': 'object',
        'dietary_tags': 'object', 'cold_chain_required': 'bool',
    },
    'delivery_request_items': {
        'line_id': 'object', 'request_id': 'object', 'item_id': 'object', 'quantity': 'int64',
    },
}

def dtype_matches(actual, expected):
    actual = str(actual)
    if expected == 'datetime64':
        return actual.startswith('datetime64')
    return actual.startswith(expected.split('64')[0])

results = []
for name, expected in expected_schemas.items():
    df = tables[name]
    missing = [col for col in expected if col not in df.columns]
    wrong_dtype = [
        f"{col}({df[col].dtype} vs {dtype})"
        for col, dtype in expected.items()
        if col in df.columns and not dtype_matches(df[col].dtype, dtype)
    ]
    results.append({
        'table': name,
        'status': 'PASS' if not missing and not wrong_dtype else 'FAIL',
        'missing': ', '.join(missing) or '-',
        'wrong_dtype': ', '.join(wrong_dtype) or '-',
    })

schema_report = pd.DataFrame(results)
display(schema_report)
assert (schema_report['status'] == 'PASS').all(), 'Schema validation failed.'
print('All 9 tables passed schema validation.')
```

```python
# Golden join plus a few convenience features
clients = clients.assign(
    full_name=clients['first_name'] + ' ' + clients['last_name'],
    wheelchair_user=clients['mobility_wheelchair'],
)
drivers = drivers.assign(
    full_name=drivers['first_name'] + ' ' + drivers['last_name'],
    is_volunteer=drivers['role_type'].eq('volunteer'),
)

items_per_request = request_items.groupby('request_id')['quantity'].sum().rename('items_delivered_count')

stops_enriched = (
    stops
    .merge(requests, on=['request_id', 'client_id'], how='left', suffixes=('', '_request'))
    .merge(clients, on='client_id', how='left', suffixes=('', '_client'))
    .merge(routes, on='route_id', how='left', suffixes=('', '_route'))
    .merge(drivers, on='driver_id', how='left', suffixes=('', '_driver'))
    .merge(items_per_request, on='request_id', how='left')
)

key_cols = [
    'route_stop_id', 'route_id', 'sequence_index', 'client_id', 'full_name',
    'program_type', 'time_window_start', 'time_window_end', 'actual_arrival',
    'status', 'failure_reason', 'wheelchair_user', 'requires_two_person_team',
    'driver_id', 'is_volunteer', 'items_delivered_count',
]
stops_enriched[key_cols].head(10)
```

```python
# EDA, 2x3 grid of six plots
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

routes['route_status'].value_counts().plot(kind='bar', ax=axes[0, 0], color='steelblue')
axes[0, 0].set_title('Route Status Breakdown')
axes[0, 0].tick_params(axis='x', rotation=30)

routes['on_time_rate'].dropna().plot(kind='hist', bins=25, ax=axes[0, 1], color='seagreen', edgecolor='white')
axes[0, 1].set_title('On-Time Rate Distribution Across Routes')

requests['program_type'].value_counts().plot(kind='bar', ax=axes[0, 2], color='darkorange')
axes[0, 2].set_title('Request Volume by Program Type')
axes[0, 2].tick_params(axis='x', rotation=30)

clients['food_security_level'].value_counts().plot(kind='bar', ax=axes[1, 0], color='slateblue')
axes[1, 0].set_title('Client Food Security Level')
axes[1, 0].tick_params(axis='x', rotation=30)

driver_stops = stops_enriched.groupby('driver_id').agg(
    total=('route_stop_id', 'count'),
    no_answer=('failure_reason', lambda s: s.fillna('').eq('no_answer').sum()),
)
driver_stops['no_answer_rate'] = (driver_stops['no_answer'] / driver_stops['total']).fillna(0)
driver_stops.sort_values('no_answer_rate', ascending=False)['no_answer_rate'].plot(kind='bar', ax=axes[1, 1], color='crimson')
axes[1, 1].set_title('No-Answer Rate per Driver')
axes[1, 1].tick_params(axis='x', rotation=60)

arrival = pd.to_datetime(stops['actual_arrival'], errors='coerce')
hour_counts = arrival.dt.hour.value_counts().sort_index()
hours = list(range(8, 19))
pd.Series([hour_counts.get(h, 0) for h in hours], index=hours).plot(ax=axes[1, 2], marker='o', color='teal')
axes[1, 2].set_title('Stops per Hour of Day')
axes[1, 2].set_xticks(hours)

plt.tight_layout()
plt.show()
```

```python
# Baseline route quality scorer
route_vehicle = routes.set_index('route_id')['vehicle_id']
vehicle_lift = vehicles.set_index('vehicle_id')['wheelchair_lift']
vehicle_refrigerated = vehicles.set_index('vehicle_id')['refrigerated']

stops_enriched = stops_enriched.assign(
    assigned_vehicle_id=stops_enriched['route_id'].map(route_vehicle),
)
stops_enriched['vehicle_has_lift'] = stops_enriched['assigned_vehicle_id'].map(vehicle_lift).fillna(False)
stops_enriched['vehicle_refrigerated'] = stops_enriched['assigned_vehicle_id'].map(vehicle_refrigerated).fillna(False)

route_stops_grp = stops_enriched.groupby('route_id')
completion = route_stops_grp.apply(lambda g: g['status'].eq('completed').mean()).rename('completion_rate')
skill_match = route_stops_grp.apply(
    lambda g: ((~g['wheelchair_user'].fillna(False)) | g['vehicle_has_lift']).mean()
).rename('skill_match_rate')
cold_chain_violations = route_stops_grp.apply(
    lambda g: (g['cold_chain_required'].fillna(False) & ~g['vehicle_refrigerated']).sum()
).rename('cold_chain_violations')
max_cv = max(cold_chain_violations.max(), 1)
cold_chain_violations_norm = (cold_chain_violations / max_cv).rename('cold_chain_violations_norm')

score_df = (
    routes[['route_id', 'driver_id', 'on_time_rate', 'planned_stops', 'actual_stops']]
    .merge(completion, on='route_id', how='left')
    .merge(skill_match, on='route_id', how='left')
    .merge(cold_chain_violations, on='route_id', how='left')
    .merge(cold_chain_violations_norm, on='route_id', how='left')
    .fillna({
        'on_time_rate': 0,
        'completion_rate': 0,
        'skill_match_rate': 1,
        'cold_chain_violations': 0,
        'cold_chain_violations_norm': 0,
    })
)

score_df['quality_score'] = (
    0.4 * score_df['on_time_rate']
    + 0.3 * score_df['completion_rate']
    + 0.2 * score_df['skill_match_rate']
    - 0.1 * score_df['cold_chain_violations_norm']
)
score_df['nn_savings_estimate_pct'] = np.where(
    score_df['planned_stops'] > 0,
    (1 - score_df['quality_score']) * 15.0,
    0,
)

score_df.sort_values('quality_score').head(5)[[
    'route_id', 'driver_id', 'on_time_rate', 'completion_rate',
    'skill_match_rate', 'cold_chain_violations', 'quality_score',
    'nn_savings_estimate_pct'
]]
```

```python
# Two useful operational views: client map and mismatch counts
fig, axes = plt.subplots(1, 2, figsize=(18, 7))

for depot_id, sub in clients.groupby('home_depot_id'):
    axes[0].scatter(sub['lng'], sub['lat'], s=12, alpha=0.5, label=f'depot {depot_id}')
axes[0].scatter(depots['lng'], depots['lat'], s=220, c='black', marker='*', edgecolor='gold', linewidth=1.5, label='depot', zorder=5)
axes[0].set_title('Client Locations by Home Depot (Victoria)')
axes[0].set_xlabel('longitude')
axes[0].set_ylabel('latitude')
axes[0].legend(loc='best', fontsize=8)

failure_reason = stops_enriched['failure_reason'].fillna('')
mismatch_counts = pd.Series({
    'wheelchair client, no-lift vehicle': int((stops_enriched['wheelchair_user'].fillna(False) & ~stops_enriched['vehicle_has_lift']).sum()),
    'allergen conflict flagged': int(failure_reason.eq('allergen_conflict').sum()),
    'interpreter gap flagged': int(failure_reason.eq('language_gap').sum()),
    'dog at address, allergic driver': int(failure_reason.eq('pet_conflict').sum()),
    'two-person unavailable': int(failure_reason.eq('requires_two_person_unavailable').sum()),
    'post-closure attempt': int(failure_reason.eq('post_closure_attempt').sum()),
    'cold chain violation': int(failure_reason.eq('cold_chain_violation').sum()),
    'address not found': int(failure_reason.eq('address_not_found').sum()),
    'refused delivery': int(failure_reason.eq('refused').sum()),
    'no answer': int(failure_reason.eq('no_answer').sum()),
}).sort_values(ascending=True)

mismatch_counts.plot(kind='barh', ax=axes[1], color='tomato')
axes[1].set_title('Mismatch / Failure Incidents')
axes[1].set_xlabel('count')

plt.tight_layout()
plt.show()
```
